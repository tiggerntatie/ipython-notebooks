# Plotting Quadratic Functions with Imaginary Numbers

The rectangular plane represents an Argand Diagram with real components on the horizontal axis and imaginary components on the vertical axis.

We can also represent complex numbers using luminoscity and hue. Using this method, a linear function like $f(x)=x$ is represented by the following plot:

```python
%matplotlib inline
from sympy.mpmath import cplot

cplot(lambda z: z, [-4,4],[-4,4], points=10000)
```

Notice that pure positive imaginary values are green, negative imaginary are purple, positive real are red and negative real are blue. The closer to the origin, the darker. The output of the function is zero (black) when the input is zero (origin).

Now plot $f(x)=x^2-1$:

```python
cplot(lambda z: z**2-1, [-4,4],[-4,4], points=10000)
```

Now plot $f(x)=x^2+1$:

```python
cplot(lambda z: z**2+1, [-4,4],[-4,4], points=10000)
```
