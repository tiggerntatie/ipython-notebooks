### [19] Domain restrictions, etc. of a rational function ###

$$ h(x)= \frac {2x^2-8} {x^2-x-6} = \frac {2(x^2-4)} {(x-3)(x+2)} = \frac {2(x+2)(x-2)}{(x-3)(x+2)} $$

Common factor in the numerator and denominator is (x+2), so a hole exists at $x=-2$. Canceling this common factor leaves you with a rational expression that defines the shape of the function:
    
$$ \frac {2(x-2)} {x-3} $$

This remaining curve is also undefined at $x=3$, which defines the vertical asymptote. Substitute the x coordinate of the hole to find the hole's coordinate and note that the function approaches a value of +2 as x approaches positive or negative infinity. Summarizing:

**Hole at (-2, 1.6) or (-2, 8/5), vertical asymptote at $x=3$ and horizontal (end behavior) asymptote at $y=2$**.

### [20] Graph and find intercepts ###

Evaluate $ \frac {2(x-2)} {x-3} $ at x=0 to find the y intercept: **(0, 4/3)**. Solve $0 = \frac {2(x-2)} {x-3} $ for x to find the x intercept:  **(2,0)**.

```python
X1 = linspace(-8,2.999) 
X2 = linspace(3.001,8)
h = lambda x : 2*(x-2)/(x-3)
Y1 = [h(x) for x in X1]   # evaluate the function at x values in X1
Y2 = [h(x) for x in X2]   # ditto for X2
plot(X1,Y1)               # plot Y vs X for X1 and X2
plot(X2,Y2)
plot([0],[4/3.],'.')      # plot some features of the function: y intercept
plot([2],[0],'.')         # x intercept
plot([-2],[1.6],'o')      # the hole location
xlim(xmin=-8, xmax=8)
ylim(ymin=-7,ymax=7)
vlines([3],-100,100, linestyles='dashed')    # throw in the asymptotes
hlines([2],-100,100, linestyles = 'dashed')
grid()
show()
```

### [34] Finding Angles from Area###

Set up the area equation, then solve for the angle:

$$40  = \frac 1 2 20 \cdot 5 \sin C$$

$$C \approx \sin^{-1} \left( \frac {2 \cdot 40} {20 \cdot 5} \right) = 53°$$

In general there are **two** angles that have the same sine (the angle and its supplement), so we can generate a second angle that will satisfy the equation by computing the supplement of the angle returned by the $\sin^{-1}$ function:

$$C \approx 180° - 53° = 127°$$

```python
from math import asin
C = degrees(asin(2*40/20/5))
print("C: {0:.0f}° and {1:.0f}°".format(C, 180-C))
```
