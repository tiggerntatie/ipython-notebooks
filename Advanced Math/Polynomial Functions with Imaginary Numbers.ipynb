# Plotting Polynomial Functions with Imaginary Numbers

The rectangular plane represents an Argand Diagram with real components on the horizontal axis and imaginary components on the vertical axis.

We can also represent complex numbers using luminosity and hue. Using this method, a linear function like $f(x)=x$ is represented by the following plot:

```python
%matplotlib inline
from mpmath import cplot, plot

def rcplot(func, xlim=[-4,4], ylim=[-4,4]):
    cplot(func, xlim, ylim, points=10000)
    plot(func, xlim)


rcplot(lambda z: z, [-4,4],[-4,4])
```

Notice that pure positive imaginary values are green, negative imaginary are purple, positive real are red and negative real are blue. The closer to the origin, the darker. The output of the function is zero (black) when the input is zero (origin).

Now plot $f(x)=x^2-1$:

```python
rcplot(lambda z: z**2-1, [-4,4],[-4,4])
```

Now plot $f(x)=x^2+1$:

```python
rcplot(lambda z: z**2+1, [-4,4],[-4,4])
```

Where are the zeros of the function?

```python
rcplot(lambda z: (z-1)*(z-3)*(z**2+1), [-1,4],[-4,4])
```

Zeros are 1, 3, +i and -1, only one local minimum.

```python
rcplot(lambda z: (z-1)*(z-3)*(z**2+0.05), [-1,4],[-0.5,0.5])
```

```python
rcplot(lambda z: (z-1)*(z-3)*(z-5)*(z-7), [0,8], [-0.5,0.5])
```
